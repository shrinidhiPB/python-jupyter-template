### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Generate data
    x = np.linspace(0, 10, 100)  # X values from 0 to 10
    y = x**2  # Y values as the square of X

    # Create the plot
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='y = x^2', color='b', linestyle='-', linewidth=2)

    # Customizing the plot
    plt.title('Square Function Plot', fontsize=14)
    plt.xlabel('X values', fontsize=12)
    plt.ylabel('Y values (X squared)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
