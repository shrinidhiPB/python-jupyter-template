### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Create a figure with a 2x2 layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # First subplot - Line Chart
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    axes[0, 0].plot(x, y, color='blue', linestyle='-', linewidth=2)
    axes[0, 0].set_title('Line Chart')
    axes[0, 0].set_xlabel('X values')
    axes[0, 0].set_ylabel('Y = sin(X)')
    axes[0, 0].grid(True, linestyle='--', alpha=0.6)

    # Second subplot - Bar Chart
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axes[0, 1].set_title('Bar Chart')
    axes[0, 1].set_xlabel('Categories')
    axes[0, 1].set_ylabel('Values')
    axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

    # Third subplot - Scatter Plot
    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    axes[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
    axes[1, 0].set_title('Scatter Plot')
    axes[1, 0].set_xlabel('X values')
    axes[1, 0].set_ylabel('Y values')
    axes[1, 0].grid(True, linestyle='--', alpha=0.6)

    # Fourth subplot - Pie Chart
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    colors = ['blue', 'red', 'green', 'orange']
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)
    axes[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
